```python
"""
4) Use nested loops to create the following:
# # # # # # # #
# # # # # # # #
# # # # # # # #
# # # # # # # #
# # # # # # # #
# # # # # # # #
# # # # # # # #
# # # # # # # #
"""
for i in range(8):
    for j in range(8):
        print("#", end = " ")
    print()
```
