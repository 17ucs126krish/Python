```python
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]
```

```python
#1) Sort the list and find the min and max age

ages.sort()
```

```python
print(ages)
```

```python
#2) Add the min age and the max age again to the list
min(ages)
```

```python
max(ages)
```

```python
#3) Find the average age (sum of all items divided by their number )

average_age = sum(ages)/len(ages)

average_age
```

```python
import statistics as st

avg_age = st.mean(ages)
avg_age
```

```python
#4) Find the median age (one middle item or two middle items divided by two)
m_age = st.median(ages)
m_age
```

```python
min_age = min(ages)
max_age = max(ages)

```

```python
#5) Compare the value of (min - average) and (max - average), use abs() method

min_diff = abs(min_age - avg_age)
max_diff = abs(max_age - avg_age)

print(f"Min difference:{min_diff}")
print(f"Max difference: {max_diff}")
```
