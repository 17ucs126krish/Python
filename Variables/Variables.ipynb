```python
'''
13) The radius of a circle is 30 meters:
     -> Calculate the area of a circle and assign the value to a variable name of area_of_circle.
     -> Calculate the circumference of a circle and assign the value to a variable name of circum_of_circle.
     -> Take radius as user input and calculate the area.
'''
#Area of circle:
radius = int(input("Enter radius of a circle:"))
pi = 3.14
def area_of_circle(pi, radius):
    return print("Area of circle:", pi * (radius ** 2))

area_of_circle(pi, radius)
```

```python
#Circumference of the circle

def circum_of_circle(pi, radius):
    return print("Circumference of a circle:", 2 * pi * radius)

circum_of_circle(pi, radius)
```
